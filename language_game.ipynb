```python
# -*- coding: utf-8 -*-
"""–ò–≥—Ä–∞: –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏ –ø–æ—Ä—è–¥–æ–∫ —Å–ª–æ–≤"""

import random
import pandas as pd
from datetime import datetime

# –°–°–´–õ–ö–ò –ù–ê –§–ê–ô–õ–´ –í GITHUB
parallel_texts_url = 'https://raw.githubusercontent.com/dzaytseva0610/language-game-data/refs/heads/main/data/parallel_texts.csv'
user_answers_url = 'https://raw.githubusercontent.com/dzaytseva0610/language-game-data/refs/heads/main/data/user_answers.csv'

print("üì• –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ –∏–∑ GitHub...")

try:
    # –ó–∞–≥—Ä—É–∂–∞–µ–º –ø–∞—Ä–∞–ª–ª–µ–ª—å–Ω—ã–µ —Ç–µ–∫—Å—Ç—ã
    df_texts = pd.read_csv(parallel_texts_url)
    print("‚úÖ –¢–µ–∫—Å—Ç—ã –∑–∞–≥—Ä—É–∂–µ–Ω—ã!")

    # –ü—ã—Ç–∞–µ–º—Å—è –∑–∞–≥—Ä—É–∑–∏—Ç—å –∏—Å—Ç–æ—Ä–∏—é –æ—Ç–≤–µ—Ç–æ–≤
    df_answers = pd.read_csv(user_answers_url)
    print("‚úÖ –ò—Å—Ç–æ—Ä–∏—è –æ—Ç–≤–µ—Ç–æ–≤ –∑–∞–≥—Ä—É–∂–µ–Ω–∞!")

except Exception as e:
    print(f"‚ùå –û—à–∏–±–∫–∞: {e}")
    print("‚ö†Ô∏è  –ü—Ä–æ–≤–µ—Ä—å —Å—Å—ã–ª–∫–∏ –∏ —Å–≤–æ–π –ª–æ–≥–∏–Ω GitHub!")

# üéÆ –ò–ì–†–ê
def play_game():
    # –í—ã–±–∏—Ä–∞–µ–º —Å–ª—É—á–∞–π–Ω–æ–µ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–µ
    row = df_texts.sample(1).iloc[0]
    text_id = row['id']
    original_sentence = row['text_en']
    translation = row['text_ru']

    # –ü–µ—Ä–µ–º–µ—à–∏–≤–∞–µ–º —Å–ª–æ–≤–∞
    words = original_sentence.split()
    shuffled_words = words.copy()
    random.shuffle(shuffled_words)
    shuffled_sentence = " ".join(shuffled_words)

    print(f"\nüéØ –ó–ê–î–ê–ù–ò–ï {text_id}")
    print(f"üá∑üá∫ –ü–µ—Ä–µ–≤–æ–¥: {translation}")
    print(f"üîÄ –ü–µ—Ä–µ–º–µ—à–∞–Ω–Ω—ã–µ —Å–ª–æ–≤–∞: [{shuffled_sentence}]")
    print("---" * 10)

    # –û—Ç–≤–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
    user_input = input("–¢–≤–æ–π –æ—Ç–≤–µ—Ç: ").strip()

    # –ü—Ä–æ–≤–µ—Ä–∫–∞
    is_correct = (user_input.lower() == original_sentence.lower())

    if is_correct:
        print("üéâ –ü—Ä–∞–≤–∏–ª—å–Ω–æ! –¢—ã –º–æ–ª–æ–¥–µ—Ü!")
    else:
        print(f"üòû –ü—Ä–∞–≤–∏–ª—å–Ω–æ –±—ã–ª–æ: '{original_sentence}'")

    # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
    new_answer = {
        'user_id': 'student_1',
        'text_id': text_id,
        'shuffled_sentence': shuffled_sentence,
        'user_answer': user_input,
        'is_correct': is_correct,
        'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }

    # –î–æ–±–∞–≤–ª—è–µ–º –≤ –∏—Å—Ç–æ—Ä–∏—é
    global df_answers
    df_answers = pd.concat([df_answers, pd.DataFrame([new_answer])], ignore_index=True)

    print(f"üìä –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {len(df_answers[df_answers['is_correct'] == True])} –∏–∑ {len(df_answers)} –ø—Ä–∞–≤–∏–ª—å–Ω—ã—Ö")

# üöÄ –ó–ê–ü–£–°–ö
print("\n" + "="*50)
print("üéÆ –ò–ì–†–ê: –í–æ—Å—Å—Ç–∞–Ω–æ–≤–∏ –ø–æ—Ä—è–¥–æ–∫ —Å–ª–æ–≤!")
print("="*50)

# –ò–≥—Ä–∞–µ–º 3 —Ä–∞–∑–∞
for i in range(3):
    play_game()
    if i < 2:
        continue_play = input("\n–ü—Ä–æ–¥–æ–ª–∂–∏—Ç—å? (–¥/–Ω): ").lower()
        if continue_play != '–¥':
            break

print("\nüèÅ –ò–≥—Ä–∞ –æ–∫–æ–Ω—á–µ–Ω–∞! –¢–≤–æ—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:")
print(df_answers[['text_id', 'is_correct']])
```
